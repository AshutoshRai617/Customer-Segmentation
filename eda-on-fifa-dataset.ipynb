```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.isna().sum()
```

```python
def fun_convert(df_value):
    
    if(pd.isna(df_value))!= True:
        try:
            value = float(df_value[1:-1])
            suffix = df_value[-1]

            if suffix == 'M':
                value = value * 1000000
            elif suffix == 'K':
                value = value * 1000
        except ValueError:
            value = 0
        return value
    else:
        return df_value

df['Release Clause']=df['Release Clause'].apply(fun_convert)
```
